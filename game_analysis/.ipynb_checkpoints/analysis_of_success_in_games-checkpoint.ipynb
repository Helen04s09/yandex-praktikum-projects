## Подготовка данных к анализу.

```python
#заменим названия столбцов
data = data.rename(columns = {'Name' : 'name', 'Platform' : 'platform', 'Genre': 'genre', 'NA_sales': 'na_sales', 'EU_sales': 'eu_sales', 'JP_sales': 'jp_sales', 'Other_sales': 'other_sales', 'Critic_Score': 'critic_score', 'User_Score': 'user_score', 'Rating' : 'rating', 'Year_of_Release' : 'year_of_release'})
```

<b>Выведем на экран первые пять строк таблицы:

```python
data.head()
```

```python
#для определения пропущенных значений используем метод isna()
data.isna().sum()
```

```python
#выведем % пропусков от общего числа строк
data.isna().mean() * 100
```

### Вывод :
* Вывела общее количество пропусков и % пропусков от общего числа строк. % пропусков довольно большой, однако мы не имеем достоверной информации для их заполнения. Считаю, что заполнять данные по среднему или медианному значению в данном случае не корректно.

```python
#для дальнейшего использования данных, изменим их тип на целочисленный:
data['year_of_release'] = data['year_of_release'].astype('Int64')
```

```python
#для дальнейшего использования данных, изменим их тип на вещественный:
data['critic_score'] = data['critic_score'].astype('float64')
```

```python
#проверим строки на начличие дубликатов
data.duplicated().sum()
```

```python
#изучим уникальные значения столбца:
data['critic_score'].unique()
```

```python
#изучим уникальные значения столбца:
data['user_score'].unique()
```

```python
#поскольку рейтинг игры нам не известен, заменим значение 'tbd' на nan
data['user_score'] = data['user_score'].replace('tbd', np.nan).astype('float64')
```

```python
#проверим результат замены данных:
data['user_score'].unique()
```

```python
#изучим уникальные значения столбца:
data['rating'].unique()
```

```python
#изучим уникальные значения столбца:
data['platform'].unique()
```

```python
#посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец
data['total_sales'] = data[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)
```

```python
#посмотрим сколько игр выпускалось в разные годы:
data['year_of_release'].value_counts()
```

### Вывод: 
* С целью подготовки данных к анализу, привела название столбцов к нижнему регистру, проверила данные на наличие пропусков и дубликатов, изучила уникальные значения столбцов, произвела замену типа данных в столбцах year_of_release, critic_score и user_score; посчитала суммарные продажи и посмотрела на распределение игр по годам - стремительный рост выпуска новых игр отмечается с 1994 года.

### Выберем актуальный период для дальнейших расчетов.

#### Посмотрим распределение продаж по платформам в период с 2007 года, когда наблюдался активный рост выпуска новых игр.

```python
data[data['year_of_release'] >= 2007].pivot_table(index= 'year_of_release',columns = 'platform', values= 'total_sales', aggfunc='sum').plot(grid=True, figsize=(18, 7))
plt.title('Распределение продаж по платформам')
```

##### Вывод:
* В 2009 на пике своих продаж находится Wii платформа, далее отмечается спад ее продаж и активный рост платформы X360 в 2010 году. В 2011 году лидирующие позиции занимает платформа PS3, а к 2012 году продажи падают практически по всем платформам. К 2013 году продажи по платформам вновь начинают расти, а к 2014 вновь начинают стремительно падать. В 2015 году лидирует платформа PS4, на втором месте после нее платформа XOne.

#### Поскольку, судя по графикам, актуальный период  для каждой платформы в среднем составляет 3 года, для дальнейших расчетов и прогноза будем считать актуальным - период с 2013 года.

```python
data_new = data[data['year_of_release'] >= 2013]
```

```python
data_new.pivot_table(index= 'year_of_release',columns = 'platform', values= 'total_sales', aggfunc='sum').plot(grid=True, figsize=(14, 5))
plt.title('Распределение продаж по платформам')
```

##### Вывод:
* В 2013 году отмечается падение продаж двух топовых платформ - PS3 и X360, одновременно с этим начинают стремительный рост новые лидеры - PS4 и XOne. С 2015 года наблюдаеися снижение продаж практически по всем платформам.

#### Выберем несколько потенциально прибыльных платформ.

* Судя по графику наиболее потенциальными платформами в актуальном периоде являются следующие платформы - PS4, XOne и 3DS.

```python
data_new.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum').loc[[2015, 2016]]
```

#### Построим график «ящик с усами» по глобальным продажам игр в разбивке по платформам.

```python
#построим диаграмму размаха для всех платформ в актуальном периоде
import warnings
warnings.filterwarnings('ignore')
data_new.boxplot(column='total_sales', by='platform', figsize=(14,9))
plt.xlabel("Распределение продаж игр по платформам")
plt.ylim(0, 2.5)
plt.show()
```

##### Вывод:

В данном случае, я считаю, что выбросы не являются ошибками в данных, а означают, что на каких-то платформах есть явные хитовые игры с огромными продажами (точнее их явно больше, чем на других). При этом, медианное значение за актуальный период, может говорить о том, что платформа популярна по сравнению с другими, а значит более вероятно, что она имеет потенциал. 

###  Посмотрим как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков.

#### Построим диаграмму рассеяния для визуализации зависимости между отзывами и продажами для наиболее потенциальной платформы PS4 в актуальном периоде.

```python
platform_ps4 = data_new.query('platform == "PS4"')
```

```python
#посмотрим на зависимость между продажами и отзывами критиков
pivot_table_critic_score = platform_ps4.pivot_table(index='critic_score', values='total_sales', aggfunc='sum')
```

```python
#для визуализации зависимости, построим диаграмму рассеяния
pivot_table_critic_score.reset_index().plot(x='critic_score', y='total_sales', kind='scatter')
plt.title('Зависимость продаж от оценки критиков')
plt.xlabel('Оценка критиков') 
plt.ylabel('Миллионы проданных копий')
plt.show()
```

```python
#определим коэффициент корреляции Пирсона
platform_ps4[['total_sales', 'critic_score']].corr()
```

#### Вывод: 
* наблюдается положительная зависимость продаж от отзывов критиков. Чем выше балл по отзывам критиков, тем больше число, проданных копий.

```python
#посмотрим на зависимость между продажами и отзывами пользователей
pivot_table_user_score = platform_ps4.pivot_table(index='user_score', values='total_sales', aggfunc='sum')
```

```python
#для визуализации зависимости, построим диаграмму рассеяния
pivot_table_user_score.reset_index().plot(x='user_score', y='total_sales', kind='scatter')
plt.title('Зависимость продаж от оценки пользователей')
plt.xlabel('Оценка пользователей') 
plt.ylabel('Миллионы проданных копий')
plt.show()
```

```python
#определим коэффициент корреляции Пирсона
platform_ps4[['total_sales', 'user_score']].corr()
```

#### Вывод:
* Коэффициент корреляции Пирсона в данном случае отрицательный. Взаимозависимость между данными параметрами есть, но очень незначительная, причем максимальные значения продаж наблюдаются при средних оценках  пользователей.

## Проверим гипотезы

### Сформулируем гипотезы:
* H0 - Средние пользовательские рейтинги платформ Xbox One и PC  не различаются
* H1 - Средние пользовательские рейтинги платформ Xbox One и PC  различаются

```python
data_new.head()
```

```python
#создадим срез данных по платформе XOne
XOne = data_new[data_new['platform'] == 'XOne']['user_score']
```

```python
XOne.describe()
```

```python
XOne.var()
```

```python
#создадим срез данных по платформе PC
PC = data_new[data_new['platform'] == 'PC']['user_score']
```

```python
PC.describe()
```

```python
PC.var()
```

<b>Вывод: Средние пользовательские рейтинги платформ практически не различаютя, однако дисперсия по платформе `PC` больше. Проведем t-тест для проверки гипотезы.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(XOne, PC, equal_var = False, nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<b>Вывод: результат теста дает основание предполагать, что нулевая гипотеза верна.

### Сформулируем гипотезы:
* H0 - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не различаются
* H1 - Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») различаются

```python
#создадим срез данных по жанру Action
Action = data_new[data_new['genre'] == 'Action']['user_score']
```

```python
data_new['genre'].unique()
```

```python
Action.describe()
```

```python
Action.var()
```

```python
#создадим срез данных по жанру Sports
Sports = data_new[data_new['genre'] == 'Sports']['user_score']
```

```python
Sports.describe()
```

```python
Sports.var()
```

<b>Вывод: средние пользовательские рейтинги платформ  и дисперсия различаютя. Проведем t-тест для проверки гипотезы.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(Action, Sports, equal_var = False, nan_policy='omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

<b>Вывод: результат теста дает основание предполагать, что нулевая гипотеза не верна. Таким образом, средние пользовательские рейтинги жанров Action  и Sports различаются.
