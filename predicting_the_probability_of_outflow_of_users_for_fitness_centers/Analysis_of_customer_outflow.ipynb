## Анализ оттока клиентов и подготовка плана действий по их удержанию

### Цель: 
<p><b> 
- научиться прогнозировать вероятность оттока (на уровне следующего месяца) для каждого клиента;
- сформировать типичные портреты клиентов: выделить несколько наиболее ярких групп и охарактеризовать их основные свойства;
- проанализировать основные признаки, наиболее сильно влияющие на отток;
- сформулировать основные выводы и разработать рекомендации по повышению качества работы с клиентами:

 1) выделить целевые группы клиентов;
    
 2) предложить меры по снижению оттока;
    
 3) определить другие особенности взаимодействия с клиентами.</b></p>
    
<p><b>Описание данных:</b><br>
    <b>Данные клиента за предыдущий до проверки факта оттока месяц:</b></p>
<ul>
    <li><code>gender</code>— пол</li>
    <li><code>Near_Location</code> — проживание или работа в районе, где находится фитнес-центр</li>
    <li><code>Partner</code> —  сотрудник компании-партнёра клуба (сотрудничество с компаниями, чьи сотрудники могут получать скидки на абонемент — в таком случае фитнес-центр хранит информацию о работодателе клиента);</li>
    <li><code>Promo_friends</code> — факт первоначальной записи в рамках акции «приведи друга» (использовал промо-код от знакомого при оплате первого абонемента)</li>
    <li><code>Phone</code> — наличие контактного телефона</li>
    <li><code>Age</code> - возраст</li>
    <li><code>Lifetime</code> - время с момента первого обращения в фитнес-центр (в месяцах)</li>
   
</ul>
<p><b>Информация на основе журнала посещений, покупок и информация о текущем статусе абонемента клиента:</b></p>

<ul>    
    <li><code>Contract_period</code> — длительность текущего действующего абонемента (месяц, 6 месяцев, год)</li>
    <li><code>Month_to_end_contract</code> — срок до окончания текущего действующего абонемента (в месяцах)</li>
    <li><code>Group_visits</code> - факт посещения групповых занятий</li>
    <li><code>Avg_class_frequency_total</code> — средняя частота посещений в неделю за все время с начала действия абонемента</li>
    <li><code>Avg_class_frequency_current_month</code> — средняя частота посещений в неделю за предыдущий месяц</li>
    <li><code>Avg_additional_charges_total</code> — суммарная выручка от других услуг фитнес-центра: кафе, спорттовары, косметический и массажный салон</li>
    <li><code>Churn</code> — факт оттока в текущем месяце</li>
</ul>

##  Построим модель прогнозирования оттока клиентов

<b>Для построения модели логистической регрессии с использованием библиотеки «liblinear», регуляризация l2 применяется по умолчанию.

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = data.drop('churn', axis=1)
y = data['churn']
```

```python
# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
# обучим StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)

# Преобразуем обучающий и валидационные наборы данных
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

```python
# зададим алгоритм для нашей модели
model = LogisticRegression(solver='liblinear', random_state=0)

# обучим модель
model.fit(X_train_st, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы

probabilities = model.predict_proba(X_test_st)[:, 1]

# бинарный прогноз
predictions = model.predict(X_test_st)

# выведем все изученные метрики
print('Метрики для логистической регрессии:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))
```

<b>Для проверки полученного результата, удалим из датафрейма два из четырех признаков с высокой корреляцией, чтобы исключить  мультиколлинеарность.

```python
data.drop(['month_to_end_contract', 'avg_class_frequency_current_month'], axis = 1, inplace = True)
```

```python
# зададим алгоритм для нашей модели
model = LogisticRegression(random_state=0)

# обучим модель
model.fit(X_train_st, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы

probabilities = model.predict_proba(X_test_st)[:, 1]

# бинарный прогноз
predictions = model.predict(X_test_st)

# выведем все изученные метрики
print('Метрики для логистической регрессии:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))
```

<b>Вывод: удивительно, но результат не изменился.</b>

```python
# зададим алгоритм для новой модели на основе алгоритма случайного леса
tree_model = RandomForestClassifier(n_estimators = 100, random_state=0)

# обучим модель
tree_model.fit(X_train_st, y_train)

# воспользуемся уже обученной моделью, чтобы сделать прогнозы
tree_predictions = tree_model.predict(X_test_st)
tree_probabilities = tree_model.predict_proba(X_test_st)[:,1]

# выведем все изученные метрики
print('Метрики для дерева принятия решения:')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, tree_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, tree_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, tree_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, tree_predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))
```

<b>Вывод:

* метрика <code>precision_score</code> - точность ( говорит, какая доля прогнозов относительно "1" класса верна. То есть смотрим долю правильных ответов только среди целевого класса). Чем ближе значение метрики к единице, тем лучше. Данная метрика у логистической регрессии выше, чем у случайного леса.
* метрика <code>recall_score</code> - полнота ( показывает, сколько реальных объектов "1" класса мы смогли обнаружить с помощью модели). Чем ближе значение метрики к единице, тем лучше. Данная метрика у логистической регрессии выше, чем у случайного леса.
* метрика <code>f1_score</code>  - сводная метрика, учитывающая баланс между метриками precision и recall. Чем ближе значение метрики к единице, тем лучше.Данная метрика у логистической регрессии выше, чем у случайного леса.

## Проведем кластеризацию клиентов

```python
# Отложим в сторону столбец с оттоком и проведем кластеризацию объектов
X = data.drop(['churn'], axis = 1)
```

```python
# обязательная стандартизация данных перед работой с алгоритмами
scaler = StandardScaler()
x_sc = scaler.fit_transform(X)

# обучаем модель, задаём число кластеров, равное 5, и фиксируем значение random_state для воспроизводимости результата
km = KMeans(n_clusters=5, random_state=0)

# прогнозируем кластеры для наблюдений (алгоритм присваивает им номера от 0 до 4
labels = km.fit_predict(x_sc)

# сохраняем метки кластера в поле датасета
data['cluster_km'] = labels

# для вывода графика
linked = linkage(x_sc, method = 'ward')
plt.figure(figsize=(16, 9))
dendrogram(linked, orientation = 'top')
plt.title('Кластеризация клиентов')
plt.show()









```

```python
# выводим статистику по средним значениям наших признаков по кластеру
data.groupby(['cluster_km']).agg('mean').T
```

<b>Вывод: кластеры 0 и 4 - самые надежные, так как значение <code>churn</code> ближе к нулю, значит меньше оттока. Кластеры 3 и 2  - это клиенты с оттоком.</b>

```python
# Построим распределения признаков для кластеров с помощью гистограммы
for col in data.columns:
    data.groupby('cluster_km')[col].hist()
    plt.legend(data['cluster_km'].unique())
    plt.xlabel("Значения признака")
    plt.ylabel("Количество")
    plt.title(col)
    plt.show()
```

```python
# Построим распределения признаков для кластеров, используя boxplot
for column in data.columns:
    fig, ax = plt.subplots()
    sns.boxplot(x = 'cluster_km', y = column, data = data)
    ax.set_title(column)
    plt.show()
```

<b>Вывод:
* клиенты с оттоком не живут или не работают в районе, где находится фитнес-центр, а постоянные клиенты - наоборот.
* почему то на графиках, надежным кластерам соответствуют признаки непостоянных клиентов - я не могу понять, где я ошиблась..
</b>

```python
# посчитаем долю оттока для каждого полученного кластера
round(data.groupby('cluster_km').agg({'churn': 'mean'}), 2).sort_values('churn', ascending=False)
```

<b>Вывод: по доле оттока, также наиболее надежными являются 0 и 4 кластеры.</b>

### основные признаки, наиболее сильно влияющие на отток:
* расположение фитнес-центра( постоянными становятся клиенты, которые живут или работают рядом с фитнес-центром)
* срок до окончания текущего действующего абонемента( постоянными становятся клиенты, которые имеют действующий абонемент на 6 мес и более)
* сотрудник компании-партнёра клуба( постоянными становятся клиенты, которые  являются сотрудниками партнеров фитнес-центра)


### рекомендации по повышению качества работы с клиентами:


1) целевые группы клиентов: соответствуют типичному портрету постоянного клиента фитнес-центра
    
2) меры по снижению оттока: расширить партнескую программу фитнес-клуба, провести акцию по скидке на покупку абонементов длительностью 6 мес и более, а также запустить акцию приведи друга - получи скидку на покупку первого абонемента.
    
3) особенности взаимодействия с клиентами: возможно предрставить скидку на покупку нового абонемента для постоянных клиентов клуба, чтобы удержать своих клиентов.
