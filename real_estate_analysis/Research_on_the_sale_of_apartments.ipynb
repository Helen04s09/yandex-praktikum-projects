### Расчитаем стоимость квадратного метра

```python
data['price_metr'] = (data['last_price'] / data['total_area']).round().astype('Int64')
```

```python
data['price_metr'].head()
```

### Категоризируем этаж квартиры; варианты — первый, последний, другой этаж квартиры

```python
def floor_definition(data):
    if data['floor'] == 1:
        return 'первый'
    elif (data['floors_total'] == data['floor']) & (data['floor'] != 1):
        return 'последний'
    else:
        return 'другой'
data['floor_cat'] = data.apply(floor_definition, axis = 1)
```

### Расчитаем соотношение жилой и общей площади, а также отношение площади кухни к общей.


```python
#Соотношение жилой площади и общей
data['k_leave_area'] = (data['living_area'] / data['total_area'] * 100).round()
```

```python
#определим медианное соотношение жилой площади и общей
data['k_leave_area'].median()
```

```python
#заполним пропуски в столбце living_area, используя медианное соотношение жилой площади и общей
data['living_area'] = data['living_area'].fillna(data['total_area'] *57  / 100)
```

```python
data['living_area'].head(15)
```

```python
#Отношение площади кухни к общей
data['kitchen_area_k_total'] = (data['kitchen_area'] / data['total_area'] * 100).round()
```

```python
#определим медианное отношение площади кухни к общей
data['kitchen_area_k_total'].median()
```

```python
#заполним пропуски в столбце kitchen_area, используя медианное отношение площади кухни к общей
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area'] *17  / 100)
```

```python
data['kitchen_area'].head(15)
```

## Общий вывод

* Обычно, продажа квартиры занимает около двух месяцев. Стоимость квартиры зависит от ее площади - чем больше площадь, тем ниже стоимость квадратного метра. При этом взаимозависимость между данными параметрами наиболее выражена в центральном сегменте. Так стоимость квадратного метра в 4-х комнатной квартире на 26% ниже, чем у однокомнатных квартир по всему городу, и на 71% в центральном сегменте. Квартиры на первом этаже имеют наименьшую стоимость. Cтоимость квадратного метра таких квартир в центральном сегменте на 14% ниже, по сравнению с другими этажами и на 5% по сравнению с квартирами, расположенными на последних этажах. По всему городу на 16% и 4,5% соответственно. Также отмечается взаимозависимость между стоимостью квадратного метра и удаленностью от центра, причем с увеличением расстояния от центра - стоимость квартиры уменьшается. А вот зависимости стоимости квартиры от даты размещения практически нет, кроме естесственного ежегодного роста цен, в том числе и на квадратный метр квартиры.
